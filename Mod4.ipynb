```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
thispage = requests.get('https://phoenix.craigslist.org/search/gms?sale_date=2022-02-04')
bs = BeautifulSoup(thispage.text,'html.parser')
latest = bs.find(id='search-results')
```

```python
def grabarticle(thisurl):
  thispage = requests.get(thisurl)
  bs = BeautifulSoup(thispage.text,'html.parser')
  body = bs.find(id='postingbody')
  return {'url':thisurl,'body':body}
```

```python
list = []
```

```python
links = latest.find_all('a')
for eachlink in links:
  full_url = eachlink['href']
  if 'tool' in full_url:
      print('getting',full_url)
      story = grabarticle(full_url)
      sleep(15)
      list.append(story)
```
